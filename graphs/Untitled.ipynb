```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

data = pd.read_csv("tacho_measurement2.csv")

plt.figure(figsize=(40, 10))
plt.step([*range(0, len(data['rot_per_s']))], data['rot_per_s'])

plt.grid()

array = data['rot_per_s'].to_numpy().copy()

for i in range(1, array.size):
    array[i] = 0.2*array[i-1] + 0.8*array[i]

plt.step([*range(len(array))], array)

plt.figure(figsize=(40, 10))
plt.step([*range(0, len(data['rot_per_s']))], data['rot_per_s'])

plt.grid()

array = data['rot_per_s'].to_numpy().copy()

for i in range(1, array.size):
    array[i] = 0.8*array[i-1] + 0.2*array[i]

plt.step([*range(len(array))], array)

plt.figure(figsize=(40, 10))

plt.step([*range(0, len(data['rot_per_s']))], data['rot_per_s'])

plt.grid()

array = data['rot_per_s'].to_numpy().copy()

for i in range(1, array.size):
    array[i] = 0.9*array[i-1] + 0.1*array[i]

plt.step([*range(len(array))], array)
```
